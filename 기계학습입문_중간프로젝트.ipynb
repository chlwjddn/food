### 2) 5대범죄 발생건수 (분자)
- 변수명 : total_teen_crime
- 데이터 : https://www.data.go.kr/data/3075889/fileData.do


```python
import pandas as pd

teen_crime_list = []

for year in range(12,22):
  index = "20" + str(year)
  teen_crime = pd.read_csv(f'청소년범죄_{year}.csv', header=0, encoding='euc-kr')
# 경찰서별로 데이터가 나와있기 때문에, 각 경찰서가 속한 '구'로 다시 인덱싱하는 작업입니다
  teen_crime['구'] = ['중구', '종로구', '중구', '서대문구', '종로구', '용산구', '성북구', '동대문구', '마포구', '영등포구', '성동구', '동작구', '광진구', '은평구', '강북구', '금천구', '중랑구', '강남구', '관악구', '강서구', '강동구', '성북구', '구로구', '서초구', '양천구', '송파구', '노원구', '서초구', '은평구', '도봉구', '강남구']
  teen_crime = teen_crime.groupby('구').sum()
  teen_crime[index] = teen_crime.iloc[:, :5].sum(axis=1)
  teen_crime_list.append(teen_crime[index])
  
total_teen_crime = pd.concat(teen_crime_list,axis = 1)
total_teen_crime
```

```python
# 휴 세로로 다시 이어붙이기 성공...
total_teen_crime = total_teen_crime.stack().reset_index()
print(type(total_teen_crime))
total_teen_crime
```

```python
total_table = pd.DataFrame() # 종속변수!

total_table['teen_crime'] = total_teen_crime[0]
# 이런 식으로 total_table에 계속 데이터 쌓아가는 형식으로 전체 데이터를 준비하면 될 것 같아요!
print(type(total_table))
print(total_table)
```

- 변수명 : total_police_num

- 데이터 : 
https://data.seoul.go.kr/dataList/224/S/2/datasetView.do

```python
import pandas as pd

police_num = pd.read_csv('경찰서_수.csv', skiprows= [1,2,28], header=0 , usecols = list(range(1,62)))
police_num = police_num.sort_values(by='자치구별(2)', ascending=True)
police_num.set_index('자치구별(2)', inplace=True)
#police_num = pd.DataFrame(police_num).fillna(0)
police_num = police_num.replace("-",0)
police_num = police_num.replace("1",1)
police_num = police_num.replace("2",2)
total_police_num = pd.DataFrame()

for year in range(12,22):
  police_per_year = police_num.iloc[:, 6*year - 71 :6*year - 65]
  total_police_num[f'20_{str(year)}'] = police_per_year.sum(axis=1)
  #print(police_per_year)이걸로 오류 찾아냄 = str 인식 못하는 데이터 직접 수정함!

total_police_num
```

```python
# 세로로 이어붙이는 작업 마찬가지.
total_police_num = total_police_num.stack().reset_index()
print(type(total_police_num))
total_police_num
```

## 로그변환 (2차)

```python
import numpy as np

total_table['Teen_Crime'] = np.log1p(total_table['teen_crime']) # 종속변수!

#total_table['Street_Light'] = np.log1p(total_table['Street_Light'])

total_table['One_Parent'] = np.log1p(total_table['One_Parent'])

total_table['Recipient_Basic'] = np.log1p(total_table['Recipient_Basic'])

total_table.head(10)
```

## 세트분할 (1차)

```python
from sklearn.model_selection import train_test_split
#train = total_table[['Family_Violence', 'Street_Light', 'One_Parent', 'Police_Station']].to_numpy() # 다중공선성 위험으로 인해 X5는 제외!
train = total_table[['Family_Violence', 'Street_Light', 'One_Parent', 'Police_Station', 'Recipient_Basic']].to_numpy() 
target = total_table['teen_crime'].to_numpy()
# input(X)과 target(Y) 데이터를 , 훈련/테스트 세트로 분리!
train_X, test_X, train_Y, test_Y = train_test_split(train, target, test_size=0.25)
```

# ▶스케일 전처리

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_X)
train_X_scaled = ss.transform(train_X) # 훈련세트 표준화
test_X_scaled = ss.transform(test_X) # 그에 맞게 테스트세트도 표준화!
```
